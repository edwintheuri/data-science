**Boolean Arrays**

**Counting number entries in the arrays eg True**


```python
y = random.randint(10, size=(2,2))
print (y)
print(np.count_nonzero(y > 2)) 
```

```python
y = random.randint(10, size=(2,2))
print (y)
print(np.sum(y > 2, axis= 1)) #axis each row
```

**Checking quick if values are true, can use any or all**


```python
y = random.randint(10, size=(2,2))
print (y)
print(np.any(y > 2, axis= 1))  #h
```

```python
y = random.randint(10, size=(2,2))
print (y)
print(np.all(y > 2, axis= 1)) 
```

**Boolen bitwise operators**

& - and 
| or a
~ used with  and (& ) or (|) meaning not
^ neither 

```python
y = random.randint(10, size=(5,7))
print (y)
print("")
print(np.sum(( y < 2) & (y == 0) , axis= 1)) 
```

```python
y = random.randint(10, size=(5,7))
print (y)
print("")
print(np.sum(( y < 2) | (y == 0) , axis= 1)) 
```

```python
y = random.randint(10, size=(5,7))
print (y)
print("")
print(np.sum(~( y > 2) & (y == 0) , axis= 1)) 
```

```python
y = random.randint(10, size=(5,7))
print (y)
print("")
print(np.sum(( y < 2) ^ (y == 0) , axis= 1)) 
```

**Masking**

- Means Indexing on selected values that have meet a certain condition 

```python
y = random.randint(10, size=(5,7))
print (y)
print("")
y[y<5] #returns a MASKED 1D array of all the values in y that are less than 5.
```

## Structured Arrays

- Makes it easy to refer to values either by name or index

```python
#sample data

name = ["alice", "bob", "charlie", "dave", "eve"]
age = [25, 30, 35, 40, 45]
weight = [150, 160, 170, 180, 190]
```

```python
#compound data type

data =np.zeros(5, dtype={"names":("name", "age", "weight"),
                          "formats": ("U10", "i4", "f8")})
data
```

- U10 means max length 10 

- i4 means 4 byres (32 bit ) Integer

- f8 means 4 byres (64 bit ) float

```python
#POPULATING THE COMPOUND DATA TYPE

data["name"] = name
data["age"] = age   
data["weight"] = weight
data
```

```python
data["name"]
```

```python
older = data[data["age"] > 30]   #prints the rows of data where the age is greater than 30
print(older)
```

```python
bob_only = data[data ["name"] == "bob"] #printing the row of data where the name is "bob"
print (bob_only)
```
