#### Create a function that will determine whether a number is positive or negative.

```python
def checker(a):
    if a > 0:
        print(f"{a} is Positive")
    if a < 0:
        print(f"{a} is Negative")
checker(2)
```

#### Write functions to calculate the area of different shapes like square (area_of_square(side_length)), rectangle (area_of_rectangle(length, width)), and circle (area_of_circle(radius)).


```python
def square (a):
    length = a * 2
    print(f"The square of {a} is {length}")
square(2)


```

```python
def rectangle(l,w):
    measurements = l * w
    print(f"The area of rectangle with length {l}, Width {w} is {measurements}")
rectangle(2,3)
```

```python
import math
def circle(r):
    area=math.pi*(r**2)
    print(f"The area of circle with radius {r} is {area}")
circle(4)
```

#### Write a function called calculate_grade that takes a score as input and returns the corresponding letter grade (A, B, C, D, F) based on a grading scale (e.g., A: 90-100, B: 80-89, etc.).

```python
def grade(a):
    if a >= 90 and a <= 100:
        print(f"You got an A,  {a}")
    elif a >= 80 and a <=89:
        print(f"You got an B,  {a}")
    elif a >= 70 and a <=79:
        print(f"You got an C,  {a}")
    elif a >= 60 and a <=69:
        print(f"You got an D,  {a}")
    elif a >= 0 and a <=59:
        print(f"You got an F,  {a}")
    else:
        ("Enter a number")
grade(65)
```
