#### Write a program that generates a random number between 1 and 100. The user has to guess the number in a limited number of attempts. Provide feedback after each guess (higher, lower, or correct).

```python
import random  

randomnumber = random.randrange(1,100) 
guesslimit = 5
print("This program generates a random number between 1-100 and gives you only 5 guess")

while guesslimit:
    userinput = int(input("Enter your guess between 1 and 100 : "))
    if userinput > randomnumber:
        print(f"{userinput} is Higher")
    elif userinput < randomnumber:
        print(f"{userinput} is Lower")
    else:
        print(f"{userinput} Correct")
    guesslimit -= 1
print(f"You have reached the limit, {guesslimit} guess left")



```

#### Write a program that allows the user to enter two numbers and an operator (+, -, *, /) and perform the corresponding calculation. Use conditional statements to handle different operators.

```python
input1 = int(input("Enter first number: "))
input2 = int(input("Enter second number: "))
operatorr=input("Enter +  - * / to perform calculations")
if operatorr == "+":
    sum = input1+input2
    print(f"The sum of {input1} and {input2} = {sum}")
elif operatorr == "-":
    diff = input1-input2
    print(f"The subtration of {input1} and {input2} = {diff}")
elif operatorr == "*":
    multi = input1*input2
    print(f"The multiplication of {input1} and {input2} = {multi}")
elif operatorr == "/":
    div = input1/input2
    print(f"The division of {input1} and {input2} = {div}")
else:
    print("Follow intructions")

 


```

#### Write a program that uses a for loop to print the multiplication table of a number entered by the user (up to a specified number of multiples).

```python
multinput = int(input("Enter a number to print multiplication table : "))
if multinput == 1:
    print(f"{multinput}* {multinput} is {multinput}")
else :
    for num in range(1, multinput):
        multable = num * multinput
        print(f"{multinput} x {num} is {multable}")
```

#### Write a program that uses a for loop to calculate the sum of numbers from 1 to a given number entered by the user.

```python
suminput = int(input("Enter a number to calculate sumation exclude 1 : "))

for summ in range(1, suminput):
    sum = suminput + summ
    print(f"{suminput} + {summ} is {sum}")
```
