#### 1. Create a program that acts as a simple calculator. It should take two numbers and an operator (+, -, *, /) as input and perform the corresponding arithmetic operation. Add comments to explain each part of the code.

```python
print ("-"*10)

num1=input("num1, no decimals:\n")

num1_init=int(num1) 

op = input("Add,Subtract,multiple,divide:\n")

num2=input("num2, no decimals:\n")

num2_init=int(num2) 

if op == "+" :  
    print(f"The sum of {num1_init} + {num2_init} is sum {num1_init+num2_init}")
elif op == "-":
    print(num1_init+num2_init)
elif op == "*":
    print(num1_init+num2_init)
elif op == "/":
    print(num1_init+num2_init)    
else:
    print("please enter +, *, /, -")
print ("-"*10)
```
