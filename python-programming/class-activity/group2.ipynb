--How to Achieve Abstraction in Python

Unlike Java, Python does not have built-in abstract classes by default. We must import the abc module (Abstract Base Classes).

The Three Rules of Abstraction:

1. Inheritance: You must inherit from ABC.
2. Decorator: You must use @abstractmethod on methods you want to enforce.
3. No Instantiation: You cannot create an object from an Abstract Class. If you try, Python will throw an error.

```python
### in this example car is an abstract concept , you can't buy a generic car , must buy a specific brand
###
#Base Class

from abc import ABC, abstractmethod

class Car(ABC):
    @abstractmethod
    def mileage (self):
        pass

class Tesla(Car):
    def mileage(self):
        print("This car mileage is 30kmph (Electric)")

class Suzuki(Car):
    def mileage(self):
        print("This car mileage is 25kmph (Petrol)")

class REnaut(Car):
    def mileage(self):
        print("This car mileage is 27kmph (Hybrid)")

class Duster(Car):
    def mileage(self):
        print("This car mileage is 24kmph (Diesel)")

T = Duster()
S= Tesla()
R = REnaut()
Su = Suzuki()

T.mileage()
S.mileage()
R.mileage()
Su.mileage()
```

```python
class BankAccount:
    def __init__(self):
        self.balance = 1000
acc = BankAccount()
acc.balance = -900000
print(acc.balance)
```

```python
class Bank_account:
    def __init__(self,balance):
        self.__balance = balance # private vairable encapsulation, accessed inside the class
    
    def Deposit(self,amount):
        self.__balance += amount
    
    def Withdraw(self,amount):
        if amount <= self.__balance:
            self.__balance -= amount #subtract money if enough balance
        else:
            print("Insufficient funds")
    def Get_balance(self):
        return self.__balance
```

```python
acc = Bank_account(1000)
acc.Deposit(500)
acc.Withdraw(300)
print(acc.Get_balance())
```

```python
class Student:
    def __init__(self): 
        self.__grade = 0
    def Set_grade(self, Grade):
        if 0 <= Grade <= 100:
            self.__grade = Grade
        else:
            print("Invalid Code")
    def Get_grade(self):
        return self.__grade
    
#create object

S1 = Student()
S1.Set_grade(85)

#modification 

S1.__grade = 1000
print(S1.Get_grade())
```
