## Inheritance


```python

class Animal:


    def __init__ (self, Name, Age):
        self.Name = Name 
        self.Age = Age
        self.is_alive = True

    def Eat (self):
        print(f"{self.Name} is eating")
    
    def Sleep(self):
        print(f"{self.Name} is sleeping")
    
    def __str__(self):
        return f"{self.__class__.__name__} : {self.Name}, Age{self.Age}"
    
    """ Child classes now """
    
class Dog(Animal):
    def __init__ (self , Name , Age , Breed):
        super().__init__(Name, Age) #calls parent
        self.Breed = Breed  #new attribute
        
    def  Bark (self):
        print(f"{self.Name} says : woof !")

    def  Fetch (self, Item):
        print(f"{self.Name} fetch an {Item}!")

class Cat(Animal):
    def __init__ (self , Name , Age , Is_indoor):
        super().__init__(Name, Age)
        self.Is_indoor = Is_indoor
        
    def  Purr (self):
        print(f"{self.Name} purrs !") 

    def  sleep (self):               #overiding parent class with new behaviour
        super().sleep()              # still calls parent version first
        print(f"..... {self.Name} sleeps 16 hours straight") 

""" Multi Instance """

class GuideDog(Dog):
    def __init__(self,Name, Age, Breed, Owner):
        super().__init__(Name, Age, Breed)
        self.Owner = Owner
    def Guide(self):
        print(f"{self.Name} carefully guides {self.Owner} across the street")

        
```
