#### REGEX Continuation 

```python
import re
text = "number is 0712548129"
results = re.search(r"\d{3}",text)
print(text[::3],results.group())
result_replace = re.sub(r"\d", "#" , text)
result_replace2 = re.sub(r"07\d{8}", r"07#######" , text)
result_replace3 = re.sub(r"(07\d{2})(\d{4}\d{2})", r"\1###\2" , text)

#matching text 
match = re.match(r"number", text)
if match:
    print(f"Its a match  {match.group()}")
else:
    print("None")

print(result_replace)
print(result_replace2)
print(result_replace3)
```

'\d' -  digits

'\w' -  letters and numbers 

'\W' - not a word character 

'\s' - space 

'S' -  non-space

'.' - any character

'+' - 1 or more 

'*' -  0 or more

'?' - optional

'^' - start

## Database


### What is a database 


organized collection of information stored in a computer system 

### Importance of a database

- secure, can require user to enter username and password
- indexing , stored in a structed way , has improved readability
- scalability, can store thousands of data , and users can access the database simultaneous.

## Characteristics 

- automicity , incase of a power surge or blackout, data that was in progress cancels out , "all or nothing "

- scalability, can accomodated billions of data without breaking 

-  data independence , you can operate on the file structure or moving

- data intergrity , possible to enforce rules to ensure accuracy, like only allowing , negatives in a column or ensure column accepts only emails. 

## Types

- Relational databases 
- no sql
- specialized databases

### Relational database

- structured:  tables , rows and columns 
- relationship : tables are linked via unique keys , like a table called customer , can link orders using customer_id
- language : uses sql , modify and define data 
-  best use case : where data has strict structure and consistency , like a financial system , inventory system etc 
- popular engine : mysql (powers the web), postresql (enterprise), sqlite (protype)


### NO sql

- structutred : collection of json , no fixed tables or columns 
- relationship: flexible, no unique keys needed 
-  best use case: social media feeds, real time data analysis etc
- popular engine : mongodb
